2. Generate at random 3 matrices of size $3 \times 3$ and fill each entry with a random integer chosen from -10 to 10. Use an existing
package to compute the inverse of each matrix if it exists. Include in your report:
    - Each of the 3 matrices
    - The inverses computed with the package
    - Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
# this way the numbers don't change everytime it re-runs
np.random.seed(42) 
```

```python
# generating the 3 matrices
matrix_1 = np.random.randint(-10, 10, size=(3,3)) 
matrix_2 = np.random.randint(-10, 10, size=(3,3)) 
matrix_3 = np.random.randint(-10, 10, size=(3,3)) 
```

```python
matrix_1, matrix_2, matrix_3
```

```python
# computes the inverse of each matrix 
inverse_1 = np.linalg.inv(matrix_1)
inverse_2 = np.linalg.inv(matrix_2)
inverse_3 = np.linalg.inv(matrix_3)
```

```python
inverse_1, inverse_2, inverse_3
```

 Compute the product of each matrix with its inverse to check that you obtain the identity matrix.

```python
product_1 = np.dot(matrix_1, inverse_1) 
```

```python
product_1 
```

```python
product_2 = np.dot(matrix_2, inverse_2) 
```

```python
product_2
```

```python
product_3 = np.dot(matrix_3, inverse_3) 
```

```python
product_3
```
