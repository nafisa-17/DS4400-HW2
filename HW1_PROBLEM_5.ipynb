```python
import pandas as pd
import numpy as np 
```

```python
# upload dataframe and get rid of columns that are not needed
kc_house_data = pd.read_csv("kc_house_data.csv")
kc_house_data = kc_house_data.drop(columns=["id", "date", "zipcode"]) 
```

### Question 1

```python
# For each feature, write code to compute the average value, 
# the min and max values, as well as its variance.

def stats(df):
    dict_1 = {}
    for column_name, column_values in df.items():
        if column_values.dtype in ["int64", "float64"]:
             average_score = float(column_values.mean())
             min_score = column_values.min()
             max_score = column_values.max()
             variance_score = column_values.var()
             dict_1[column_name] = {'average': average_score,
                                    'min': min_score,
                                    'max': max_score,
                                    'variance': variance_score}
    return dict_1 

```

```python
# applies my function
kc_house_data_1 = stats(kc_house_data)
```

```python
# the lowest and the highest average
averages = {col: vals["average"] for col, vals in kc_house_data_1.items()}

min_col = min(averages, key=averages.get)
max_col = max(averages, key=averages.get)

min_avg = averages[min_col]
max_avg = averages[max_col]

print(min_col, min_avg)
print(max_col, max_avg)

```

```python
# Finds the highest and lowest variances 
variances = {columns: values["variance"] for columns, values in kc_house_data_1.items()}
min_col = min(variances, key=variances.get)
max_col = max(variances, key=variances.get)

min_var = variances[min_col]
max_var = variances[max_col]

print(min_col, min_var)
print(max_col,max_var )

```
